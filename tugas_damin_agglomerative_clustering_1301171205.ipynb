```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
points = np.array([[10,4],[12,25],[15,66],[20,23],[45,24],[33,53],[21,67],[55,54],[32,78],[12,45],[96,65],[55,77],[68,23],[98,4],[49,15],[60,23],[44,12],[50,64],[9,7],[85,5],[4,4],[63,55],[3,20],[18,30],[40,44],[54,14],[3,25],[76,26],[12,48],[42,63]])
# print(points)
plt.scatter(points[:,0], points[:,1])
```

```python
dendrogram = sch.dendrogram(sch.linkage(points, method='ward'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage='ward') 
y_hc = hc.fit_predict(points)
```

```python
plt.scatter(points[y_hc == 0,0], points[y_hc == 0,1], s = 40, c='red')
plt.scatter(points[y_hc == 1,0], points[y_hc == 1,1], s = 40, c='green')
plt.scatter(points[y_hc == 2,0], points[y_hc == 2,1], s = 40, c='blue')
```
